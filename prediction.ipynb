Import library yang diperlukkan.

```python
import numpy as np
import pandas as pd
from keras.layers import Dense
from keras.models import Sequential
import tensorflow as tf
```

Memuat data csv

```python
dataset = pd.read_csv('Datatest.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

Membagi data menjadi data latih dan data uji

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

Normalisasi/FeatureScaling data menggunakan StandardScaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train.reshape(-1, 1))
y_test = sc_y.transform(y_test.reshape(-1, 1))
```

```python
print(X_train)
```

```python
print(X_test)
```

Membuat model Artificial Neural Network

```python
model = Sequential()
model.add(Dense(units=5, activation='relu'))
model.add(Dense(units=5,activation='relu'))
model.add(Dense(units=5,activation='relu'))
model.add(Dense(units=1 ,activation='sigmoid'))
```

```python
model.compile(loss='mean_absolute_error', optimizer='adam')
```

Melatih ANN

```python
model.fit(X_train, y_train, epochs=2000,  validation_data=(X_test, y_test), verbose=2)
```

Memprediksi jumlah kendaraan bermotor di Indonesia pada tahun 2022, dengan memasukkan data total jumlah penduduk di Indonesia dan laju pertumbuhan penduduk pada tahun 2022.

```python
prediksi_2022=model.predict(sc.transform([[275773800, 1.17]]))
print(prediksi_2022)
prediksi_2022=sc_y.inverse_transform(prediksi_2022)
print(f"Prediksi jumlah kendaraan bermotor di Indonesia pada tahun 2022 adalah : {int(prediksi_2022)}")
```
